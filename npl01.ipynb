```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
tokenized_sents = [word_tokenize(i) for i in data["STORY"]]
print(tokenized_sents)
```

```python
flattened = []
for sublist in tokenized_sents:
    for val in sublist:
        flattened.append(val)
print(flattened)
```

```python
Vocab=[]
for item in flattened:
    if not item in Vocab:
        Vocab.append(item)
print(Vocab)
```
